## Local Inference on GPU
Model page: https://huggingface.co/mahmoudalyosify/Horus-OSINT

⚠️ If the generated code snippets do not work, please open an issue on either the [model repo](https://huggingface.co/mahmoudalyosify/Horus-OSINT)
			and/or on [huggingface.js](https://github.com/huggingface/huggingface.js/blob/main/packages/tasks/src/model-libraries-snippets.ts) 🙏

```python
!pip install llama-cpp-python

from llama_cpp import Llama

llm = Llama.from_pretrained(
	repo_id="mahmoudalyosify/Horus-OSINT",
	filename="llama-3-8b-instruct.Q4_K_M.gguf",
)
```

```python
llm.create_chat_completion(
	messages = [
		{
			"role": "user",
			"content": "What is the capital of France?"
		}
	]
)
```

```python
# 1. Reload the model with GPU offloading enabled
llm_gpu = Llama.from_pretrained(
    repo_id="mahmoudalyosify/Horus-OSINT",
    filename="llama-3-8b-instruct.Q4_K_M.gguf",
    n_gpu_layers=-1, # <--- The Magic Word to use the T4 GPU!
    n_ctx=2048,      # Expand context window for longer reports
)

# 2. The OSINT/Intelligence Prompt
response = llm_gpu.create_chat_completion(
    messages = [
        {
            "role": "system",
            "content": "You are Horus-OSINT, an advanced intelligence analyst AI."
        },
        {
            "role": "user",
            "content": "What was the geopolitical situation in India in 2018, and what type of attack occurred there according to OSINT records?"
        }
    ],
    max_tokens=500,
    temperature=0.2 # Low temperature for factual, consistent reporting
)

# 3. Print the output beautifully
print(response['choices'][0]['message']['content'])
```
